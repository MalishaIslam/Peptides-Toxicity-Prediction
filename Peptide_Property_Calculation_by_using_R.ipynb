```r
library("Peptides")
```

```r
library("reticulate")
```

```r
pep_data <- read.csv("peptide.csv")

pep_data
```

```r
n_c_list = list()

for (row in 1:nrow(pep_data)){
  seq <- pep_data[row, "Sequence"]
  #print(seq)
  n_c <- charge(seq,pH= 7, pKscale= "EMBOSS")
  n_c_list <- c(n_c_list, n_c)
}
```

```r
length(n_c_list)
```

```r
n_c_list
```

```r
df <- as.data.frame(array(unlist(n_c_list),dim=c(3864, 1)))
df
```

```r
write.csv(df,"/content/netchargedata.csv", col.names = 'net_charge')
```

# Extract Dipeptide Composition

```r
#install.packages("protr")
```

```r
library("protr")
```

```r
pep_data <- read.csv("/content/peptide.csv")

pep_data
```

```r
dc_list = list()

for (row in 1:nrow(pep_data)){
  seq <- pep_data[row, "Sequence"]
  #print(seq)
  dc_list <- c(dc_list, seq)
}
```

```r
length(dc_list)
```

```r
x2 <- t(sapply(dc_list, extractDC))
colnames(x2)[colnames(x2) == "NA"] <- "NA_dipeptide"
x2
```

```r
write.csv(x2,"/content/dcdata.csv")
```
